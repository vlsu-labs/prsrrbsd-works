```python
import pandas as pd

FILE_PATH = 'train.csv'

MAIN_DF = pd.read_csv(FILE_PATH)

MAIN_DF
```

Описание признаков:

**PassengerId** - id пассажира

**Survived** бинарная переменная: выжил пассажирил (1) или нет (0)

**Pclass** - класс пассажира

**Name** - имя пассажира

**Sex** - пол пассажира

**Age** - возраст пассажира

**SibSp** - количество родственников (братьев, сестер, супругов) пассажира на борту

**Parch** - количество родственников (родителей / детей) пассажира на борту

**Ticket** - номер билета

**Fare** - тариф (стоимость билета)

**Cabin** - номер кабины

**Embarked** - порт, в котором пассажир сел на борт (C - Cherbourg, S - Southampton, Q = Queenstown)

Задание: 

1) Посчитайте количество пассажиров первого класса, которые сели на борт в Саутгемптоне.

2) Солько пассажиров третьего класса, которые путешествовали в компании 2 или более родственников (братьев / сестер / супругов)?

3) Сколько в среднем стоил билет первого класса?

```python
TARGET_COLUMNS = ['Pclass', 'Embarked']
TARGET_VALUES = [1, 'S']

df = MAIN_DF.copy()

locs = None

for column, value in zip(TARGET_COLUMNS, TARGET_VALUES):
    new_loc = df[column] == value
    # print(new_loc)
    if locs is None:
        locs = new_loc
        continue
    
    locs = locs & new_loc

result1 = df.loc[locs]
print(f"Result1: {len(result1.index)}")
# print(result1.to_string())

loc = (df['SibSp'] >= 2) & (df['Pclass'] == 3)
result2 = df.loc[loc]
print(f"Result2: {len(result2.index)}")
# print(result2.to_string())


result3 = df.loc[df['Pclass'] == 1]
print(f"Result2: {result3['Fare'].mean()}")
```

Задание: отличается ли распределение выживших и не выживших среди мужчин и женщин?

```python
df = MAIN_DF.copy()

df = df.sort_values('Survived')

result = df.groupby('Sex')['Survived'].value_counts()

dead_stat = result['male'][0] / result['female'][0]
live_stat = result['male'][1] / result['female'][1]

print(dead_stat)
print(live_stat)
# Да
```
